```python
from nba_api.stats.endpoints import playerprofilev2
from nba_api.stats.static import players

import pandas as pd
```

```python
last_name = 'Gilgeous-Alexander'
```

```python
player_df = pd.DataFrame(players.get_active_players())
```

```python
def find_id(last_name):
    player = player_df[player_df['last_name'] == last_name]
    try:
        player_id = player['id'].iloc[0]
        return player_id
    except: 
        print(player)


player_id = find_id(last_name)
```

```python
player_profile_list = playerprofilev2.PlayerProfileV2(player_id).get_data_frames()
```
